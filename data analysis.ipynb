```python
import pandas as pd 
import matplotlib.pyplot as plt 
from pyecharts.charts import *
import pyecharts.options as opts
from pyecharts.commons.utils import JsCode
from pyecharts.globals import SymbolType
```

```python
path = './胡润数据_民营500.xlsx' #引入生成的民营500强的数据文件
df = pd.read_excel(path,index_col=0)
df.head()
```

## 排名前100的富豪

```python
df1 = df[:100]  #获取财富前100的数据
chart = Bar( #初始化绘图模板
    init_opts=opts.InitOpts(
        theme= 'walden'
        ,width='800px'
        ,height='400px'
        ,bg_color='white'
    )
)

chart.add_xaxis(['第{}名:{}'.format(str(i+1), df1['hs_Rank_CTop500_ChaName_Cn'][i]) for i in range(len(df1['hs_Rank_CTop500_ChaName_Cn']))])

chart.add_yaxis(
    ''
    ,df1['hs_Rank_CTop500_Wealth_USD'].tolist()
)

chart.set_global_opts(
    title_opts=opts.TitleOpts(title='排名前100的富豪（$）')
    ,xaxis_opts=opts.AxisOpts(
            axislabel_opts={"rotate":25}
        )
    #配置区域缩放
    ,datazoom_opts=[opts.DataZoomOpts(
            range_start=0,  # 开始范围
            range_end=20,  # 结束范围
            type_='slider',  # slider形式
            is_zoom_lock=False,  # 锁定区域大小
            pos_left='10%',  # 设置位置
            pos_top='5%'
        )]
    ,tooltip_opts=opts.TooltipOpts(
        formatter='{b}<br>财富：{c}万'              #文本内容
    )
    ,yaxis_opts=opts.AxisOpts(min_='30') 
)
chart.render_notebook()
```

## 排名向上

```python
#此段代码获取需要处理的几个字段：排名、行业、公司、人名、财富等，选出排名上升超过50的名单，并根据排名增加幅度进行排序处理
temp = df[['hs_Rank_CTop500_ChaName_Cn','hs_Rank_CTop500_Ranking','hs_Rank_CTop500_Ranking_Change','hs_Rank_CTop500_Industry_Cn']]
temp = temp[temp['hs_Rank_CTop500_Ranking_Change'].notnull()]
temp['hs_Rank_CTop500_Ranking_Change'] = temp['hs_Rank_CTop500_Ranking_Change'].astype(int)
temp = temp.sort_values('hs_Rank_CTop500_Ranking_Change', ascending=False)
temp = temp[temp['hs_Rank_CTop500_Ranking_Change']>50].reset_index(drop=True)
temp['hs_Rank_CTop500_OldRanking'] = temp['hs_Rank_CTop500_Ranking']+temp['hs_Rank_CTop500_Ranking_Change']
temp['hs_Rank_CTop500_ChaName_Cn'] = ['{}:{}'.format(temp['hs_Rank_CTop500_ChaName_Cn'][i], temp['hs_Rank_CTop500_Industry_Cn'][i]) for i in range(len(temp['hs_Rank_CTop500_ChaName_Cn']))] 
```

```python
bar = Bar(   #初始化绘图模板
    init_opts=opts.InitOpts(
        width='800px'          #画布大小
        ,height='450px'
        ,theme='light'           #画布颜色
        ,bg_color='pink'        #背景颜色
    )
)
bar.add_xaxis(temp['hs_Rank_CTop500_ChaName_Cn'].tolist())
bar.add_yaxis('现排名'
              ,temp['hs_Rank_CTop500_Ranking'].tolist()
              ,stack='stack1'
              ,yaxis_index=0
              ,itemstyle_opts=opts.ItemStyleOpts(color='rgba(255, 66, 93, .5)')
              ,label_opts=opts.LabelOpts(
                  font_size=12
                  ,font_weight='bold'
                  ,position='insideTop'
              )
)
bar.add_yaxis('原排名'
              ,temp['hs_Rank_CTop500_OldRanking'].tolist()
              ,stack='stack1'
              ,itemstyle_opts=opts.ItemStyleOpts(color='rgba(147, 224, 255, .5)')
              ,label_opts=opts.LabelOpts(
                  font_size=12
                  ,font_weight='bold'
                  ,position='insideTop'
              )
)
bar.extend_axis(yaxis=opts.AxisOpts(is_show=False))
bar.set_global_opts(
    title_opts=opts.TitleOpts('排名上升超过50名的人（$）')
    ,xaxis_opts=opts.AxisOpts(
        
        axislabel_opts=opts.LabelOpts(  # 坐标轴标签配置
            font_size=12  # 字体大小
            ,font_weight='bold'  # 加重
            ,rotate=25
        )
    )
    ,yaxis_opts=opts.AxisOpts(is_show=False)
    ,datazoom_opts=[opts.DataZoomOpts(
                range_start=0,  # 开始范围
                range_end=20,  # 结束范围
                type_='slider',  # slider形式
                is_zoom_lock=False,  # 锁定区域大小
                pos_left='10%',  # 设置位置
                pos_top='5%'
    )]
)
line = Line()
line.add_xaxis(temp['hs_Rank_CTop500_ChaName_Cn'].tolist())
line.add_yaxis(
    '排名上升数'
    ,temp['hs_Rank_CTop500_Ranking_Change'].tolist()
    ,yaxis_index=1
    ,color='rgb(137, 175, 7)'
    ,label_opts=opts.LabelOpts(  # 标签设置
        font_size=10,   # 字体大小
        font_weight='bold',  # 加粗
    )
)

bar.overlap(line)
bar.render_notebook()
```

## 需要关注的行业-新晋500强和排名上升100名以上

```python
#此段代码获取需要处理的几个字段：排名、行业、公司、人名、财富等，选出排名上升超过100或新进排名的名单，并根据当前排名进行排序处理
temp = df[['hs_Rank_CTop500_ChaName_Cn','hs_Rank_CTop500_Ranking','hs_Rank_CTop500_Ranking_Change','hs_Rank_CTop500_Industry_Cn']]
temp = temp[temp['hs_Rank_CTop500_Ranking_Change'].notnull()]
temp['hs_Rank_CTop500_Ranking_Change'] = temp['hs_Rank_CTop500_Ranking_Change'].astype(int)
temp = temp[temp['hs_Rank_CTop500_Ranking_Change']>100].reset_index(drop=True)
temp['hs_Rank_CTop500_ChaName_Cn'] = ['{}:{}'.format(temp['hs_Rank_CTop500_ChaName_Cn'][i], temp['hs_Rank_CTop500_Industry_Cn'][i]) for i in range(len(temp['hs_Rank_CTop500_ChaName_Cn']))] 

temp2 = df[['hs_Rank_CTop500_ChaName_Cn','hs_Rank_CTop500_Ranking','hs_Rank_CTop500_Ranking_Change','hs_Rank_CTop500_Industry_Cn']]
temp2 = temp2[temp2['hs_Rank_CTop500_Ranking_Change'].isnull()]
temp2['hs_Rank_CTop500_ChaName_Cn'] = ['{}:{}'.format(temp2['hs_Rank_CTop500_ChaName_Cn'][i], temp2['hs_Rank_CTop500_Industry_Cn'][i]) for i in range(len(temp2['hs_Rank_CTop500_ChaName_Cn']))] 

v_temp = temp.append(temp2)

#temp = temp.sort_values('hs_Rank_CTop500_Ranking_Change', ascending=False)
#temp = temp.sort_values('hs_Rank_CTop500_Ranking', ascending=True)
```

```python
field_dir = {}  #根据上述数据获取行业前10的统计数据并排序
for i in v_temp['hs_Rank_CTop500_Industry_Cn']:
    if '、' in i:
        field = list(i.split('、'))
    else:
        field = [i]
    for j in field:
        if j in field_dir.keys():
            field_dir[j] += 1
        else:
            field_dir[j] = 1
field_top10 = sorted(field_dir.items(), key=lambda x: x[1], reverse=True)[:10]  

# 新建一个饼图 根据上述数据进行绘图
chart = Pie(
    init_opts=opts.InitOpts(
        theme='light',  # 主题
        width='500px',  # 画布大小
        height='500px',
        bg_color='white'
    )
)

chart.add(
    '',
    field_top10,
    radius=["30%", "45%"],  # 半径范围，内径和外径
    # 饼图、仪表盘、漏斗图: {a}（系列名称），{b}（数据项名称），{c}（数值）, {d}（百分比）
    label_opts=opts.LabelOpts(formatter="{b}: {d}%"  # 标签设置，{d}表示显示百分比
    )
)
chart.set_global_opts(
    title_opts=opts.TitleOpts(title="需要关注的行业", ),  # 标题设置  新晋500强和排名上升100名以上
    # 图例设置
    legend_opts=opts.LegendOpts(
        is_show=True,  # 是否显示图例
        pos_left='left',  # 图例距离左边边缘距离
        pos_top='5%',  # 图例距离上边边缘距离
    ),
)
chart.render_notebook()
```
